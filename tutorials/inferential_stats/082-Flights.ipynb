```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
flights = pd.read_csv('Flights.csv', header=0)
print(list(flights))
print(len(flights))
```

```python
flights.columns = [x.split(" ")[0] for x in list(flights)]
```

```python
list(flights)
```

```python
# calculate r and make a scatter plot
sns.jointplot(x="Distance", y="Cost", data=flights, kind = "reg")
```

```python
# calculating the standard deviation of a sample in order to calculate the slope of a regression line
n = len(flights)
std_y = np.std(flights["Cost"], ddof=1)
std_x = np.std(flights["Distance"], ddof=1)

r = stats.pearsonr(flights["Cost"],flights["Distance"])

print(r)
print(std_y, std_x)
```

```python
# so the slope is r(std_y/std_x)
slope = r[0]*(std_y/std_x)
print(slope)
```
